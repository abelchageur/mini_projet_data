```python
import pandas as pd
df = pd.read_excel('REGLEMENTS_CARTES_PREPAYEES_FAST_FOOD.xlsx')

df.head()
```

```python
df['Date_Règlement'] = pd.to_datetime(df['Date_Règlement'], format='%d/%m/%y')
df['Heure_Règlement'] = pd.to_datetime(df['Heure_Règlement'], format='%H:%M:%S').dt.time
df.info()
```

```python
M_Rgl = df['Montant_Rgl']
M_Verse = df['Montant_Versé']
S_cpp = df['Solde_CPP']

def convert_date_to_str(column):
    list_column = []
    for part in column:
        if isinstance(part, str):  # Si c'est déjà une chaîne (probablement un montant), on la garde
            list_column.append(part)
        elif hasattr(part, 'day') and hasattr(part, 'month'):  # Si c'est une date, on la formate
            day = part.day
            month = part.month
            convert_it = f"{day}.{month}"
            list_column.append(convert_it)
        else:
            list_column.append(str(part))  # Si ce n'est ni chaîne ni date, on le convertit en chaîne par sécurité
    return list_column

df['Montant_Rgl'] = convert_date_to_str(M_Rgl)
df['Montant_Versé'] = convert_date_to_str(M_Verse)
df['Solde_CPP'] = convert_date_to_str(S_cpp)
df.head()
```

```python
df['Montant_Rgl'] = pd.to_numeric(df['Montant_Rgl'], errors='coerce')
df['Solde_CPP'] = pd.to_numeric(df['Solde_CPP'], errors='coerce')
df.head()
```

```python
# Grouper les données par jour, semaine et mois
transactions_day = df.groupby(df['Date_Règlement'])['Montant_Rgl'].sum().reset_index()
transactions_week = df.resample('W', on ='Date_Règlement')['Montant_Rgl'].sum().reset_index()
transactions_month = df.resample('ME', on='Date_Règlement')['Montant_Rgl'].sum().reset_index()
```

```python
Q1 = df['Montant_Rgl'].quantile(0.25)
Q3 = df['Montant_Rgl'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

anomalies = df[(df['Montant_Rgl'] < lower_bound) | (df['Montant_Rgl'] > upper_bound)]
```

```python
anomalies.loc[:, 'Heure'] = df['group_by_hour']

grouped_anomalies_client = anomalies.groupby('Bénéficiaire_CPP').size().reset_index(name='Nombre_Anomalies')
grouped_anomalies_client_OR = grouped_anomalies_client.sort_values(by='Nombre_Anomalies', ascending=False)
grouped_anomalies_client_OR_10 = grouped_anomalies_client_OR.head(10)

grouped_anomalies_restaurant = anomalies.groupby('Restaurant').size().reset_index(name='Nombre_Anomalies')
grouped_anomalies_restaurant_OR = grouped_anomalies_restaurant.sort_values(by='Nombre_Anomalies', ascending=False)

grouped_anomalies_heure = anomalies.groupby('Heure').size().reset_index(name='Nombre_Anomalies')
grouped_anomalies_heure['Heure'] = grouped_anomalies_heure['Heure'].astype(str)
grouped_anomalies_heure_OR = grouped_anomalies_heure.sort_values(by='Nombre_Anomalies', ascending= False)
grouped_anomalies_heure_OR_10 = grouped_anomalies_heure_OR.head(10)
```

```python
# Bar chart for clients with the most anomalies
plt.figure(figsize=(10, 4))
plt.bar(grouped_anomalies_client_OR_10['Bénéficiaire_CPP'], 
        grouped_anomalies_client_OR_10['Nombre_Anomalies'], 
        color='skyblue', 
        width=0.6)
plt.title("Anomalies des montants des transactions par client", fontsize=13)
plt.xlabel("Les Clients", fontsize=10)
plt.ylabel("Nombre d'anomalies", fontsize=10)
plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()

# Bar chart for restaurants with the most anomalies
plt.figure(figsize=(10, 5))
plt.bar(grouped_anomalies_restaurant_OR['Restaurant'], 
        grouped_anomalies_restaurant_OR['Nombre_Anomalies'], 
        color='lightcoral', 
        width=0.6)
plt.title("Anomalies des transactions par restaurant", fontsize=13)
plt.xlabel("Les Restaurants", fontsize=10)
plt.ylabel("Nombre d'anomalies", fontsize=10)
plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Bar chart for hours with the most anomalies
plt.figure(figsize=(10, 4))
plt.bar(grouped_anomalies_heure_OR_10['Heure'], 
        grouped_anomalies_heure_OR_10['Nombre_Anomalies'], 
        color='mediumseagreen', 
        width=0.6)
plt.title("Anomalies des transactions par heure de la journée", fontsize=13)
plt.xlabel("Les Heures", fontsize=10)
plt.ylabel("Nombre d'anomalies", fontsize=10)
# plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

*Anomalies par client : Le graphique montre une grande disparité entre les clients en termes de nombre d'anomalies. Certains clients présentent un nombre significativement plus élevé d'anomalies que d'autres.

*Anomalies par restaurant : Plusieurs restaurants se distinguent avec un nombre d'anomalies supérieur à la moyenne. Cela suggère des problèmes spécifiques à ces établissements.

*Anomalies par heure de la journée : Les heures de 13h et 12h semblent être les plus sujettes aux anomalies, ce qui pourrait indiquer des problèmes liés aux pics d'activité ou aux changements de quart.

*Compréhension du graphique:

Le graphique présente une corrélation positive forte entre le solde des cartes prépayées et les dépenses totales des clients. Cela signifie qu'en général, plus le solde d'une carte prépayée est élevé, plus les dépenses associées à cette carte sont importantes.
